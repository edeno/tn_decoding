```python
import numpy as np

spike_times_sec = spike_times.index / np.timedelta64(1, 's')
```

```python
position_info['time_sec'] = position_info.index / np.timedelta64(1, 's')
```

```python
def get_position_at_spikes(spike_times_sec, position_info):
    '''
    spike_times_sec must be the times in seconds
    position_info must be modified to have a 'time_sec' column with corresponding units to spike_times_sec
    '''
    positions_at_spikes = np.empty((spike_times_sec.shape[0], 2), dtype=position_info['x_position'].dtype)

    time_ind = 0

    for i, e in enumerate(spike_times_sec):
        for j in range(time_ind, position_info['time_sec'].shape[0]):
            prev_pos_time = position_info['time_sec'][max(0, j-1)]
            pos_time = position_info['time_sec'][j]
            # if we have gone too far
            if pos_time > e:
                time = prev_pos_time
                time_ind = max(0, j-1)
                break
        # time_ind (where to index into positions table
        positions_at_spikes[i, :] = [position_info['x_position'][time_ind], position_info['y_position'][time_ind]]
    return positions_at_spikes
```

```python
import matplotlib.pyplot as plt

def plot_spatial_tuning_curve_from_positions_at_spikes(positions_at_spikes):
    fig, ax = plt.subplots()
    ax.scatter(positions_at_spikes[:,0], positions_at_spikes[:, 1])
    return fig, ax
```

```python
def plot_spikes_histogram(spike_times_sec):
    fig, ax = plt.subplots()
    ax.hist(spike_times_sec, bins = 30)
    return fig, ax
```

```python
positions_at_spikes = get_position_at_spikes(spike_times_sec, position_info)
```

```python
plot_spatial_tuning_curve_from_positions_at_spikes(positions_at_spikes)
```

```python
plot_spikes_histogram(spike_times_sec)
```

```python
figs = []

for i in range(neuron_info.index.shape[0]):
    spike_times = get_spikes_dataframe(neuron_info.index[i], ANIMALS)
    spike_times_sec = spike_times.index / np.timedelta64(1, 's')
    positions_at_spikes = get_position_at_spikes(spike_times_sec, position_info)
    fig, ax = plot_spatial_tuning_curve_from_positions_at_spikes(positions_at_spikes)
    ax.set_xlabel('x position')
    ax.set_ylabel('y position')
    fig.suptitle(f'place cell {i}')
    figs.append(fig)
    fig.savefig(f'place_cell_{i}.png')
    
    fig, ax = plot_spikes_histogram(spike_times_sec)
    fig.suptitle(f'histogram {i}')
    ax.set_xlabel('time (s)')
    ax.set_ylabel('times fired in bin')
    fig.savefig(f'histogram_{i}.png')
```
